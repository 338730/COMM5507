```python
import pandas as pd 
import numpy as np
```

```python
csv_file = "fish.csv"
gapminder = pd.read_csv(csv_file)
```

```python
print(type(gapminder))
gapminder
```

```python
from snownlp import SnowNLP
```

```python
gapminder['emo_ti'] = gapminder['caption'].apply(lambda x:SnowNLP(x).sentiments)
gapminder.head(10)
```

```python
gapminder['emo_co'] = gapminder['content'].apply(lambda x:SnowNLP(x).sentiments)
gapminder.head(10)
```

```python
gapminder['emo_ti'] .describe()
```

```python
gapminder['emo_co'] .describe()
```

```python
# Calculate the number for positive titles and negative titles
pos = 0
neg = 0
for i in gapminder['emo_ti']:
    if i >= 0.5:
        pos += 1
    else:
        neg += 1
print('The number for titles and negative titles is：')
pos,neg
```

```python
# The percentage for postive songs and negative title
import matplotlib.pyplot as plt 

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False

pie_labels='postive','negative'
plt.pie([pos,neg],labels=pie_labels,autopct='%1.1f%%',shadow=True)

plt.show()
```

```python
pos = 0
neg = 0
for i in gapminder['emo_co']:
    if i >= 0.5:
        pos += 1
    else:
        neg += 1
print('The number for titles and negative titles is：')
pos,neg
```

```python
# The percentage for postive songs and negative title
import matplotlib.pyplot as plt 

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False

pie_labels='postive','negative'
plt.pie([pos,neg],labels=pie_labels,autopct='%1.1f%%',shadow=True)

plt.show()
```

```python
import matplotlib.pyplot as plt


plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False

bins=np.arange(0,1.1,0.1)
plt.hist(gapminder['emo_ti'],bins,color='#4F94CD',alpha=0.9)
plt.xlim(0,1)
plt.xlabel('Sentiment grade')
plt.ylabel('count')
plt.title('Title Sentiment Histogram')
plt.show()
```
