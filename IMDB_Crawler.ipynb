```python
browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get('https://www.imdb.com/title/tt10919420/reviews?ref_=tt_urv')

while(1):
    temp_response=browser.page_source
    temp_soup=BeautifulSoup(temp_response,'lxml')
    load_more=temp_soup.select('.ipl-load-more__button')

    if len(load_more)==1:
        button_load_more=browser.find_element_by_class_name('ipl-load-more__button')
#         print(button_load_more)
        if button_load_more.is_displayed():
            button_load_more.click()
            time.sleep(10)
        else:
            break
    else:
        break
```

# Task 3: Search Tuniu for airtickets

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait as Wait
from selenium.webdriver.support import expected_conditions as Expect
browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get('https://www.imdb.com/title/tt10919420/reviews?ref_=tt_urv')
```

```python
depart='SHA' # shanghai
destine='BJS' # beijing
deptDate='2021-10-24'
arrvDate='2021-10-28'

# https://flight.tuniu.com/domestic/list/SZX_BJS_SRT_1_0_0/?start=2020-10-22&back=2020-10-24

browser.get('http://flight.tuniu.com/domestic/list/'+depart+'_'+destine
            +'_SRT_1_0_0?start='+deptDate+'&back='+arrvDate)
```

```python
Wait(browser, 60).until(
    Expect.text_to_be_present_in_element((By.ID, "loadingStatus"), u"共搜索")
)
```

```python
flight_items = browser.find_elements_by_class_name("flight-item")
for flight_item in flight_items:
    flight_price_row = flight_item.find_element_by_class_name("flight-item-price")
    price_element = flight_price_row.find_element_by_class_name('price')
    print(price_element.text)
```
